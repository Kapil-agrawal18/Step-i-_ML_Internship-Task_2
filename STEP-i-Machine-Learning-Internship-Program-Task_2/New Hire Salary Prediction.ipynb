# Feature Engineering

```python
# forming list of iterable tuples having 2 int dtype columns name in each tuple

from itertools import combinations

col_int = dataset.select_dtypes(exclude = 'object')

column_list = col_int.columns
interactions = list(combinations(column_list, 2))
interactions
```

```python
from sklearn.linear_model import LinearRegression

itr_dict = {}
feature_list = []
for itr in interactions:
    X = col_int.copy()
    X['int'] = X[itr[0]] * X[itr[1]]
    regression = LinearRegression()
    regression.fit(X, y_train)
    itr_dict[regression.score(X, y_train)] = itr
    
    
top_3 = sorted(itr_dict.keys(), reverse = True)
for itr in top_3:
    print(itr_dict[itr])
    feature_list.append(itr_dict[itr])
```

```python
# function to create new features in dataset

def create_features(dataset,itr_dic=feature_list):
    for itr in itr_dic:
        dataset[itr[0]+"_"+itr[1]] = dataset[itr[0]] * dataset[itr[1]]
    return dataset

dataset = create_features(dataset)
test = create_features(test)
```

# Polynomial Feature Engineering

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly_dict = {}
feature_list = []
for feature in col_int.columns:
    for p in range(2, 6):
        X = col_int.copy()
        X['poly'] = X[feature] ** p
        regressor_1 = LinearRegression()
        regressor_1.fit(X, y_train)
        poly_dict[regressor_1.score(X, y_train)] = [feature, p]
    
feature_list.append(poly_dict[max(poly_dict.keys())])
```

```python
feature_list
```

```python
# function to create new polynomial features in dataset

def create_poly_features(dataset,poly_dic=feature_list):
    for poly in poly_dic:
        dataset[poly[0]+"_"+str(poly[1])] = dataset[poly[0]] ** poly[1]
    return dataset

dataset = create_poly_features(dataset)
test = create_poly_features(test)
```

Changing Categorical data to integers

```python
def  conv_obj_columns(dataset):
    col_obj = dataset.select_dtypes(include = 'object')
    for col in col_obj.columns:
        vect = pd.get_dummies(col_obj[col])
        dataset = dataset.drop(col,axis=1)
        dataset = pd.concat([dataset,vect],axis=1)
    return dataset
        
dataset = conv_obj_columns(dataset)
test = conv_obj_columns(test)
```

Repeating Frature engineering step

```python
from itertools import combinations

column_list = dataset.columns
interactions = list(combinations(column_list, 2))
interactions
```

```python
dataset = create_features(dataset,itr_dic=interactions)
test = create_features(test,itr_dic=interactions)
```

# Standard Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()

dataset = pd.DataFrame(sc.fit_transform(dataset),columns = dataset.columns)
test = pd.DataFrame(sc.transform(test), columns = test.columns)
```

```python
X_train = dataset.iloc[:,:].values
X_test = test.iloc[:,:].values
```

# Dimension Reduction using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components =90)
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

# Train Regression Model

```python
from sklearn.linear_model import LinearRegression
reg_ = LinearRegression()
reg_.fit(X_train, y_train)
```

# Make Prediction

```python
y_pred = reg_.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print(f"r^2: {r2_score(y_test, y_pred)}")
print(f"rmse: {np.sqrt(mean_squared_error(y_test, y_pred))}")
```

```python
output = pd.DataFrame((np.concatenate((y_test.reshape(len(y_test),1), y_pred.reshape(len(y_pred),1)),1)),columns=["Actual CTC","Predicted CTC"])
```
